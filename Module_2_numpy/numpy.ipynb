## độ lệch chuẩn (đặc trưng cho độ phân tán của dữ liệu) = căn của phương sai

### phương sai: var = $\sum_{i = 0}^{n} $$\frac{(xi - \overline{x})^2}{n}$
### std = sqrt(var)

vd: matrix = (1 2
              3 4)
              
=> n = 4 (có 4 số)
$\overline{x}$ = (1+2+3+4)/n = 10/4 = 2.5
=> var = $\frac{(1 - 2.5)^2 + (2 - 2.5)^2 + (3 - 2.5)^2 + (4 - 2.5)^2}{4}$ = $\frac{2.25 + 0.25 + 0.25 + 2.25}{4}$ = 1.25
=> std = sqrt(var) = sqrt(1.25) = 1.118
             


```python
matrix = np.array([
    [1,2],
    [3,4]
])

print(matrix)

print('độ lệch chuẩn std:', matrix.std()) # = np.std(matrix)
print('phương sai var:', np.var(matrix))

matrix.std(axis = 1) #tính theo dòng 
matrix.var(axis = 1) #
```

## lấy phần giao giữa 2 mảng (các phần tử xuất hiện mảng 1 và mảng 2)

```python
arr1 = np.array([1,2,4,6,-4,1,43,2,7,8, 100])
arr2 = np.array([1,2,4,6,6,8,-2,5,2,22,199])

#lấy phần giao giữa 2 mảng (các phần tử xuất hiện mảng 1 và mảng 2) 
#và giá trị là unique
# arr3 = arr1 giao arr2
arr3 = np.intersect1d(arr1, arr2)
arr3

#lấy phần hiệu giữa 2 mảng (lấy các phần tử có trong arr1 mà ko có trong arr2 mà)
# arr4 = arr1 - arr2
arr4 = np.setdiff1d(arr1, arr2)
arr4
```
