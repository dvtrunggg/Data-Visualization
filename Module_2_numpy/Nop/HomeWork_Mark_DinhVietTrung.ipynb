```python
import numpy as np
import math
```

```python
#hàm check số x có phải là số chính phương hay không
def isPerfectSquare(x):
    sq_root = int(math.sqrt(x))
    if (sq_root * sq_root) == x:
        return True
    return False
```

### Câu 1: 
Xây dựng hàm kiểm tra số chính phương với tham số truyền vào là mảng def
kiem_tra_so_chinh_phuong(mang) để kiểm tra mảng truyền vào có phần tử nào
là số chính phương không (số chính phương là một số nguyên và có căn bậc 2 cũng là
một số nguyên. Vd : 9 là số chính phương (căn bậc 2 của 9 = 3 là một số nguyên).
Kết quả trả về mảng các giá trị True hoặc False tương ứng các phần tử trong mảng có
phải là số chính phương hay không.


```python
#hàm ktra mảng truyền vào có phần tử nào là số chính phương không
def kiem_tra_so_chinh_phuong(mang):
    tempMatrix = np.empty([mang.shape[0], mang.shape[1]], dtype = bool)
    for i in range(mang.shape[0]):
        for j in range(mang.shape[1]):
            tempMatrix[i][j] = isPerfectSquare(mang[i][j])
    return tempMatrix
```

### Câu 4

số cột), giá trị]

```python
#thông tin về vị trí của số chính phương
#vd: indexList = [[0, 1, 2], [3, 2, 1]] nghĩa là số chính phương ở vị trí matrix[0][3], matrix[1][2] và matrix[2][1]
indexList = [i.tolist() for i in np.where(check)]

resList = [] #list các phần tử là số chính phương theo định dạng : [(chỉ số dòng, chỉ số cột), giá trị]

for i in range(len(indexList[0])):
    temp = []
    temp.append((indexList[0][i], indexList[1][i]))
    temp.append(matrix[indexList[0][i]][indexList[1][i]])
    
    resList.append(temp)
    
resList
```
