```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('My_Map_Files/stations.shp')
gdf.head()
```

```python
gdf.loc[0, 'geometry']
```

```python
gdf2 = gpd.read_file('data/school_districts.geojson.txt')
gdf2.head()
```

```python
gdf2.loc[0, 'geometry']
```

```python
print(gdf2.loc[0, 'geometry'])
```

```python
gdf2.plot(column='district', figsize=(8,6))
plt.title('Nashville map')
plt.show()
```

```python
df = pd.read_csv('data/schools.csv')
df.head()
```

```python
plt.scatter(df['Longitude'], df['Latitude'], c='darkblue', marker='p')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Nashville public schools')
plt.show()
```

```python
# layer1: map
gdf2.plot(column='district', figsize=(8,6), legend=True)
# layer2: scatter
plt.scatter(df['Longitude'], df['Latitude'], c='darkblue', marker='p', edgecolor='white')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Nashville schools')
plt.show()
```

```python
# layer1: map
leg = {'title': 'District number', 'loc': 'upper left', 'bbox_to_anchor': (1, 1.03), 'ncol': 3}
gdf2.plot(column='district', figsize=(8,6), legend=True, legend_kwds=leg)
# layer2: scatter
plt.scatter(df['Longitude'], df['Latitude'], c='darkblue', marker='p', edgecolor='white')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Nashville schools')
plt.show()
```

```python
# layer1: map
leg = {'title': 'District number', 'loc': 'upper left', 'bbox_to_anchor': (1, 1.03), 'ncol': 3}
gdf2.plot(column='district', figsize=(8,6), legend=True, legend_kwds=leg, cmap='tab20')
# layer2: scatter
plt.scatter(df['Longitude'], df['Latitude'], c='darkblue', marker='p', edgecolor='white')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Nashville schools')
plt.show()
```
