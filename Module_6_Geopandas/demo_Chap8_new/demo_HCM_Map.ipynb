```python
import pandas as pd
import folium
```

```python
df_hcm = pd.read_excel('data/HCMC_location.xlsx', engine='openpyxl', sheet_name='Location')
df_hcm.head()
```

```python
# vẽ bản đồ HCM
hcm_latitude = 10.7752187
hcm_longitude = 106.6808529

hcm_map = folium.Map(location = [hcm_latitude, hcm_longitude], zoom_start = 5)
hcm_map
```

# vẽ các marker ứng với các tọa độ của các quận lên map

```python
hcm_map = folium.Map(location = [hcm_latitude, hcm_longitude], zoom_start = 12)

# khởi tạo thể hiện map
features = folium.map.FeatureGroup()

# duyệt qua các dòng dữ liệu để tạo marker

for lat, long in zip(df_hcm['Latitude'], df_hcm['Longitude']): #duyệt vĩ độ và kinh độ trong tập hợp đó
    features.add_child(
        folium.CircleMarker(
            location = [lat, long],
            radius = 5,
            color='red',
            fill=True,
            fill_color='yellow',
            fill_opacity=0.5
        )
    )
hcm_map.add_child(features)
```

```python
hcm_map = folium.Map(location = [hcm_latitude, hcm_longitude], zoom_start = 12)
# khởi tạo thể hiện map
features = folium.map.FeatureGroup()
# duyệt qua các dòng dữ liệu để tạo marker
for lat, long in zip(df_hcm['Latitude'], df_hcm['Longitude']):
    features.add_child(
        folium.Marker(
            location = [lat, long]
        )
    )
hcm_map.add_child(features)
```

### Thêm pop-up cho marker

```python
hcm_map = folium.Map(location = [hcm_latitude, hcm_longitude], zoom_start = 8)
# khởi tạo thể hiện map
features = folium.map.FeatureGroup()
# duyệt qua các dòng dữ liệu để tạo marker
for lat, long, name in zip(df_hcm['Latitude'], df_hcm['Longitude'], df_hcm['Name']):
    features.add_child(
        folium.Marker(
            location = [lat, long],
            popup=name
        )
    )
hcm_map.add_child(features)
```

### Thêm Tooltip cho marker

```python
hcm_map = folium.Map(location = [hcm_latitude, hcm_longitude], zoom_start = 12)
# khởi tạo thể hiện map
features = folium.map.FeatureGroup()
# duyệt qua các dòng dữ liệu để tạo marker
for lat, long, name in zip(df_hcm['Latitude'], df_hcm['Longitude'], df_hcm['Name']):
    features.add_child(
        folium.Marker(
            location = [lat, long],
            tooltip=name
        )
    )
hcm_map.add_child(features)
```

```python
# kết hợp popup và tooltip 
hcm_map = folium.Map(location = [hcm_latitude, hcm_longitude], zoom_start = 12)
# khởi tạo thể hiện map
features = folium.map.FeatureGroup()
# duyệt qua các dòng dữ liệu để tạo marker
for lat, long, name, population, price in zip(df_hcm['Latitude'], df_hcm['Longitude'], df_hcm['Name'], df_hcm['Population'], df_hcm['Avg_land_price']):
    features.add_child(
        folium.Marker(
            location = [lat, long],
            tooltip=name,
            popup = 'Dân số: ' + str(population) + '\nGiá BĐS: ' + str(price)
        )
    )
hcm_map.add_child(features)
```
