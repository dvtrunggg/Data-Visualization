```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/schools.csv')
df.head()
```

```python
plt.scatter(df['Longitude'], df['Latitude'], c='darkblue', marker='p')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Nashville public schools')
plt.show()
```

```python
location = [x[1:-1].split(', ') for x in df['Mapped Location']]
location
```

```python
df['lat'] = [loc[0] for loc in location]
df['long'] = [loc[1] for loc in location]
```
