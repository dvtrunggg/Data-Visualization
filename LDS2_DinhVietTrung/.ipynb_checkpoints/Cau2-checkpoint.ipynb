```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 
from PIL import Image
import numpy as np
```

Đọc dữ liệu và thực hiện chuẩn hóa (loại bỏ phần văn bản subject trước dấu : và các ký tự đặc biệt
như \r, \n) (0.5 điểm)


```python
df = pd.read_csv('data/Game_of_Thrones_arya_data.csv')
df.head(10)
```

```python
#chuẩn hóa văn bản

#chuẩn hóa tên của cột
if 'ARYA (pushing BRAN): Move!\r\n' in df.columns:
    df.rename(columns={'ARYA (pushing BRAN): Move!\r\n': 'Move!'}, inplace=True)

#chuẩn hóa các dòng 
df['Move!'] = df['Move!'].str.split(':', n=1).str.get(-1)
df.replace({r'\r\n': ''}, regex=True, inplace = True)




```

```python
df
```

Tạo biểu đồ Wordcloud

```python
text = ''
for i in df.values:
    text += str(i[0])
    
    
```

```python
my_stop_word = set(STOPWORDS)
my_stop_word.add('arya')
my_stop_word.add('stark')

wc = WordCloud(background_color = 'white', stopwords = my_stop_word).generate(text)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
wc_mask = np.array(Image.open('data/Arya.png')) 
```

```python
wc_mask1 = WordCloud(background_color = 'white', stopwords = my_stop_word, mask = wc_mask)
wc_mask1.generate(text)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(wc_mask1, interpolation = 'bilinear')  
plt.axis('off')
plt.show()
```
