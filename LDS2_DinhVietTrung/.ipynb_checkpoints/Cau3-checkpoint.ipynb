```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
```

1. Đọc dữ liệu, hiển thị thông tin chung của dữ liệu : head, tail, info, describe. (0.25 điểm)

```python
df = pd.read_csv('data/udemy_courses.csv')
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

2. Vẽ biểu đồ phân phối tần suất các khóa học theo học phí (price) (0.5 điểm)

```python
warnings.filterwarnings('ignore')
```

```python
fg, (ax1, ax2) = plt.subplots(1, 2) 
sns.distplot(df['price'], ax = ax1)
sns.distplot(df['price'], ax = ax2, kde = False)

plt.show()

```

3. Nhận xét biểu đồ trên (0.25 điểm)


```python
print('Nhận xét: Từ biểu đồ, ta thấy số liệu tập trung ở phía tay trái (phân bố lệch phải - right skewed)')
```

4. Tính hệ số tương quan giữa 2 thuộc tính num_lectures và content_duration, sau đó vẽ biểu đồ (0.5 điểm)


```python
df.corr()
print(df['num_lectures'].corr(df['content_duration']))
```

```python
pg = sns.PairGrid(data = df, vars = ['num_lectures', 'content_duration'])
pg2 = pg.map(plt.scatter)
plt.show()
```

5. Vẽ biểu đồ thể hiện sự tương quan giữa num_lectures và content_duration (mỗi hình 0.5 điểm)

```python
#fg, (ax1, ax2) = plt.subplots(1, 2) 

sns.lmplot(data=df, x='num_lectures', y='content_duration', hue = 'level')
plt.show()

sns.jointplot(data=df, x='num_lectures', y='content_duration', kind='reg', color = 'red')
plt.show()
```

6. Nhận xét biểu đồ trên (0.5 điểm)

```python
print('Nhận xét: Từ biểu đồ trên, ta thấy mối tương quan thuận (tuyến tính dương) giữa num_lectures và content_duration')
```

7. Nhóm dữ liệu level và đếm theo course_id, cho biết mỗi nhóm có bao nhiêu? (0.5 điểm). 

```python
dfGroup = df.groupby('level').course_id.count()
print(dfGroup)
```

8. Vẽ biểu đồ thể hiện dữ liệu level theo gợi ý như hình sau : (0.5 điểm)

```python
plt.figure(figsize=(8,8)) 
plt.pie(dfGroup, autopct='%.0f%%', startangle= 180, labels = dfGroup.index)
plt.legend(dfGroup.index, loc = 4)
plt.show()
```
