Xây dựng hàm kiểm tra số nguyên tố def kiem_tra_so_nguyen_to(so) để kiểm tra số truyền vào
có phải là số nguyên tố hay không (số nguyên tố là số >= 2 và chỉ chia hết cho 1 và chính nó, ví dụ :
2, 7, 13). Kết quả trả về True nếu là số nguyên tố, ngược lại trả về False

```python
def kiem_tra_so_nguyen_to(so):
    if so >= 2:
        for i in range(2, so):
            if not ( so % i ):
                return False
    else:
        return False
    return True        
```

mảng 2 chiều có kích thước 4x4 với các phần tử có giá trị phát sinh ngẫu nhiên từ 1 đến 20


```python
matrix = np.random.randint(1, 21, (4,4))
print(matrix)
```

Kiểm tra và xuất ra danh sách các phần tử trong mảng là số nguyên tố theo định dạng 
[(chỉ số dòng, chỉ số cột), giá trị phần tử]

```python
# hàm ktra các ptu trong matrix là số ngto hay ko, return về matrix True False
def Ma_Tran_Ptu_Ngto(matrix):
    tempMatrix = np.empty([matrix.shape[0], matrix.shape[1]], dtype = bool)
    
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            tempMatrix[i][j] = kiem_tra_so_nguyen_to(matrix[i][j])
    return tempMatrix
```

```python
checkMatrix = Ma_Tran_Ptu_Ngto(matrix)
indexList = [i.tolist() for i in np.where(checkMatrix)]

resList = [] #list các phần tử là số chính phương theo định dạng : [(chỉ số dòng, chỉ số cột), giá trị]

for i in range(len(indexList[0])):
    temp = []
    temp.append((indexList[0][i], indexList[1][i]))
    temp.append(matrix[indexList[0][i]][indexList[1][i]])
    
    resList.append(temp)
    
print(resList)
```

Thay thế giá trị của các phần tử không phải là số nguyên tố trong mảng bằng 0

```python
matrix = np.where(checkMatrix, matrix, 0)
print(matrix)
```
