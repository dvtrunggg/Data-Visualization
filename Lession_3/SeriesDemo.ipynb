## thêm phần tử vào series

```python
seri2 = pd.Series([-4.4,5.8,2.2,-5.4,100,5.6,2.3,5,5,2.4])

#thêm zô nè:)
seri2[11] = 100
seri2

#xóa nè
seri2 = seri2.drop(11)
seri2

#xóa trực tiếp (ko tạo bản sao => ko cần gán)
seri2.drop(9, inplace = True)
seri2

```

```python
#xóa theo index
seri2 = pd.Series([-4.4,5.8,2.2,-5.4,100,5.6,2.3,5,5,2.4])
seri2.drop(seri2.index[[2,5]], inplace = True)
seri2
```
