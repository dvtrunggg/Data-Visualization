```python
def checkChar(str):
    listChar = str.split(' ')
    shortestChar = ''
    longestChar = ''
    for i in range(len(listChar) - 1):
        if(len(listChar[i+1]) > len(listChar[i])):
            longestChar = (listChar[i+1])
            
        if(len(listChar[i+1]) < len(listChar[i])):
            shortestChar.append(listChar[i+1])
            
    return [shortestChar, longestChar]
```

```python
str = 'Write a Java program to sort an array of given integers using Quick sort Algorithm'
print(checkChar(str))
```
