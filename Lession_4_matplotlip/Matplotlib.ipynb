### Lọc ra dòng có chưa outliers trong cột cân nặng

```python
#tính ở mức 25%
Q1 = np.quantile(df['cannang'], 0.25) 
Q1
```

```python
IQR = Q3 - Q1
cannang_upper = Q3 + 1.5*IQR
cannang_upper

#vì ko có giá trị lower outliers nên ko cần tính:)
```

```python
dfOutliers = df[df['cannang'] > cannang_upper]
dfOutliers
```
