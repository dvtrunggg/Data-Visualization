```python
# Câu 4: Tạo FacetGrid của total_bill theo 'time' và chỉ định thứ tự của các cột bằng col_order
# Bạn nhận xét gì về biểu đồ vừa tạo
fg = sns.FacetGrid(data = tips,  height = 4, aspect = 2, col_order = ['Lunch', 'Dinner'], col= 'time')
fg.map(sns.boxplot, 'total_bill', order = None)

plt.show()
plt.clf()
```

```python
# Câu 5: Tạo catplot chứa point plot của giá trị 'total_bill' thay đổi theo size và tách dòng theo ‘time’
# Bạn nhận xét gì về biểu đồ vừa tạo

print(tips.head())

sns.catplot(data = tips, 
            x = 'size',  #thay đổi theo size
            y = 'total_bill', # catplot chứa point plot của giá trị 'total_bill'
            kind = 'point',  # chứa point plot
            row = 'time') #tách dòng theo ‘time’
plt.show()
plt.clf()
```

```python
# Câu 6: Tạo PairGrid với scatter plot thể hiện liên quan giữa "total_bill" và "tip" 
# Bạn nhận xét gì về biểu đồ vừa tạo
pg = sns.PairGrid(data = tips, vars = ["total_bill", "tip"])
pg2 = pg.map(plt.scatter)

plt.show()
```

```python
# Câu 7: Tạo Pairplot với scatter plot thể hiện liên quan giữa "total_bill" và "tip", sử dụng palette color = 'day' 
# Bạn nhận xét gì về biểu đồ vừa tạo
sns.pairplot(data = tips, vars = ["total_bill", "tip"],  #liên quan giữa "total_bill" và "tip"
             palette = 'bright', 
             hue = 'day', 
             kind = 'scatter', #scatter plot
            diag_kws = {'alpha': 0.9})
plt.show()
plt.clf()
```
