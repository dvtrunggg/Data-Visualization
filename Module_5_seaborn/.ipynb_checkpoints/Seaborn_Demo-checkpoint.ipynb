## Biểu đồ distplot
### Tương tự histogram nhưng nó vẽ thêm 1 đường KDE - giúp xem xét hình dạng phân phối dữ liệu

```python
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('schoolimprovement2010grants.csv', sep=',', #sep = ',' là ngăn cách giữa các data là dấu , 
                 index_col=0) # set luon index là cột đầu tiên do trong file csv đã có cột index
df.head()
```

```python
df.info()
print('có 757 dòng mà tại region: 747 non-null => có 10 dòng null tại dòng này')
```

```python
# vẽ biểu đồ distplot với tham số mặc định
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green')
plt.show()

#sns.histplot(df['Award_Amount'], color='red')
#plt.show()

##có warning thì muốn bỏ => import warning


print('từ biểu đồ => phân phối lệch phải (căn từ đỉnh của nó á)')
```

```python
# không hiển thị đường kde và đổi màu cho biểu đồ
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', kde=False)
plt.show()
```

```python
# không hiển thị phần histogram chỉ hiển thị đường kde (dương phân phối) và đổi màu cho biểu đồ
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', hist=False)
plt.show()
```

```python
# không hiển thị đường kde và đổi màu cho biểu đồ, thay đổi số lượng cột của hist
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', kde=False, bins=20) #20 cột nè:))))
plt.show()
```

## Xem mật độ dữ liệu :Distrobution Plot

```python
# hiển thị mật độ của dữ liệu
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='violet', rug=True,hist=False)
plt.show()

print('data tập trubg ở khoảng màu liên tục')
```

```python
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='r', rug=True,hist=False, kde_kws={'shade':'kde_kws'})
plt.show()
```

## Regression Plot

```python
df = pd.read_csv('insurance_premiums.csv')
df.head()
```

```python
sns.regplot(data = df, x = 'insurance_losses', y = 'premiums')
plt.show()
print('Kết luận: mối tương quan giữa insurance_losses và premiums là tương quan thuận')
```

```python
corr = df[['premiums', 'insurance_losses']].corr()

print('ktra lại: độ tương quan: corr = ', corr.insurance_losses[0])
corr
```

## Lmplot: giống như regression plot

```python
df.head(10)
```

```python
# vẽ biểu đồ lmplot với tham số mặc định
plt.figure(figsize=(8, 5))
sns.lmplot(data=df, x='insurance_losses', y='premiums')
plt.show()
```

```python
# vẽ biểu đồ lmplot với tham số hue (phân nhóm dữ liệu theo màu sắc)
sns.lmplot(data=df, x='insurance_losses', y='premiums', 
           hue='Region') #hue='Region' sẽ set các nhóm theo màu sắc
plt.show()
```

```python
# vẽ biểu đồ lmplot với tham số col (phân nhóm dữ liệu theo cột)
sns.lmplot(data=df, x='insurance_losses', y='premiums', 
           col='Region') #(phân nhóm dữ liệu theo cột) (bao nhiêu giá trị thì chia bấy nhiêu cột)
plt.show()
```

```python
# vẽ biểu đồ lmplot với tham số row (phân nhóm dữ liệu theo dòng) - đơn giản là mỗi biểu đồ 1 dòng hoy
sns.lmplot(data=df, x='insurance_losses', y='premiums', row='Region')
plt.show()
```

```python
df = sns.load_dataset('iris')
df.head()
```

## swarmplot

```python
# xem xét sepal_length theo từng loại hoa (species)

#sắp xếp các dữ liệu ko chồng, đè lên nhaooo
sns.swarmplot(data=df, x='species', y='sepal_length')
plt.show()
```

## jointplot: biểu đồ kết hợp á:)))

```python
# xem xét mối tương quan giữa petal_length và petal_width
sns.jointplot(data=df, x='petal_length', y='petal_width')
plt.show()
```

```python
#kind : "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" 
sns.jointplot(data=df, x='petal_length', y='petal_width', kind='reg', color='red')
plt.show()
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# vẽ barplot mặc định
sns.barplot(data=df, x='species', y='petal_length') #mặc định estimator=np.mean
plt.show()

##đường gạch màu đen (error bar) trên mỗi cột thể hiện độ tin cậy của dữ liệu đó
```

```python
# vẽ barplot sử dụng estimator
sns.barplot(data=df, x='species', y='petal_length', estimator=np.sum) #tính theo sum
plt.show()

sns.barplot(data=df, x='species', y='petal_length', estimator=np.max) #tính theo max
plt.show()
```

```python
# vẽ barplot mặc định
sns.barplot(data=df, x='species', y='petal_length', ci=99)
plt.show()
```

```python
# vẽ lineplot mặc định
sns.lineplot(data=df, x='species', y='petal_length')
plt.show()
```
