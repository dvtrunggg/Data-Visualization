```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

## swarmplot

```python
# xem xét sepal_length theo từng loại hoa (species)
sns.swarmplot(data=df, x='species', y='sepal_length')
plt.show()
```

## jointplot

```python
# xem xét mối tương quan giữa petal_length và petal_width
sns.jointplot(data=df, x='petal_length', y='petal_width')
plt.show()
```

```python
#kind : "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" 
sns.jointplot(data=df, x='petal_length', y='petal_width', kind='reg', color='red')
plt.show()
```
