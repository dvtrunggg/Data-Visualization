```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/schoolimprovement2010grants.csv', sep=',', index_col=0)
df.head()
```

```python
df.info()
```

```python
# vẽ biểu đồ distplot với tham số mặc định
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'])
plt.show()
```

```python
# không hiển thị đường kde và đổi màu cho biểu đồ
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', kde=False)
plt.show()
```

```python
# không hiển thị phần histogram chỉ hiển thị đường kde và đổi màu cho biểu đồ
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', hist=False)
plt.show()
```

```python
# không hiển thị đường kde và đổi màu cho biểu đồ, thay đổi số lượng cột của hist
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', kde=False, bins=20)
plt.show()
```

```python
# hiển thị mật độ của dữ liệu
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', rug=True,hist=False)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.distplot(df['Award_Amount'], color='green', rug=True,hist=False, kde_kws={'shade':'kde_kws'})
plt.show()
```
