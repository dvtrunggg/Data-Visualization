```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/insurance_premiums.csv')
df.head()
```

```python
# vẽ biểu đồ lmplot với tham số mặc định
plt.figure(figsize=(8, 5))
sns.lmplot(data=df, x='insurance_losses', y='premiums')
plt.show()
```

```python
# vẽ biểu đồ lmplot với tham số hue (phân nhóm dữ liệu theo màu sắc)
sns.lmplot(data=df, x='insurance_losses', y='premiums', hue='Region')
plt.show()
```

```python
# vẽ biểu đồ lmplot với tham số col (phân nhóm dữ liệu theo cột)
sns.lmplot(data=df, x='insurance_losses', y='premiums', col='Region')
plt.show()
```

```python
# vẽ biểu đồ lmplot với tham số row (phân nhóm dữ liệu theo dòng)
sns.lmplot(data=df, x='insurance_losses', y='premiums', row='Region')
plt.show()
```
