```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/schoolimprovement2010grants.csv', index_col=0)
df.head()
```

## barplot

```python
sns.barplot(data=df, x='Award_Amount', y='Model Selected', hue='Region')
plt.show()
```

```python
sns.barplot(data=df, x='Model Selected', y='Award_Amount')
plt.show()
```

## poitplot

```python
sns.pointplot(data=df, x='Model Selected', y='Award_Amount')
plt.show()
```

```python
# to 2 ng gch ngang cho error bar
sns.pointplot(data=df, x='Model Selected', y='Award_Amount', capsize=0.1)
plt.show()
```

## Countplot

```python
sns.countplot(data=df, x='Model Selected')
plt.show()
```

```python
sns.countplot(data=df, y='Model Selected')
plt.show()
```
