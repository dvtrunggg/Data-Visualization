```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/schoolimprovement2010grants.csv', index_col=0)
df.head()
```

## BoxPlot

```python
# xem xet Award_Amount theo Model selected
sns.boxplot(data=df, x='Award_Amount', y='Model Selected')
plt.show()
```

## Violinplot

```python
# xem xet Award_Amount theo Model selected
sns.violinplot(data=df, x='Award_Amount', y='Model Selected')
plt.show()
```

## Boxenplot

```python
# xem xet Award_Amount theo Model selected
sns.boxenplot(data=df, x='Award_Amount', y='Model Selected')
plt.show()
```
