```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# vẽ barplot mặc định
sns.barplot(data=df, x='species', y='petal_length')
plt.show()
```

```python
# vẽ barplot sử dụng estimator
sns.barplot(data=df, x='species', y='petal_length', estimator=np.sum)
plt.show()
```

```python
# vẽ barplot mặc định
sns.barplot(data=df, x='species', y='petal_length', ci=99)
plt.show()
```

```python
# vẽ lineplot mặc định
sns.lineplot(data=df, x='species', y='petal_length')
plt.show()
```
