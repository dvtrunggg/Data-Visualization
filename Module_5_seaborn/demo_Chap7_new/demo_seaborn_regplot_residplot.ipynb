```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/college_datav3.csv')
df.head()
```

## regplot

```python
# xem xÃ©t tÆ°Æ¡ng quan giá»¯a SAT_AVG_ALL vÃ  Tuition
sns.regplot(data=df, x='SAT_AVG_ALL', y='Tuition')
plt.show()
```

```python
sns.regplot(data=df, x='SAT_AVG_ALL', y='Tuition', marker='*', color='purple')
plt.show()
```

```python
# xem xÃ©t tÆ°Æ¡ng quan giá»¯a SAT_AVG_ALL vÃ  PCTPELL
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='PCTPELL', y='SAT_AVG_ALL')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='PCTPELL', y='SAT_AVG_ALL', order=2)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='PCTPELL', y='SAT_AVG_ALL', order=3)
plt.show()
```

## residplot

```python
plt.figure(figsize=(8,6))
sns.residplot(data=df, x='SAT_AVG_ALL', y='Tuition')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.residplot(data=df, x='PCTPELL', y='SAT_AVG_ALL')
plt.show()
```

## Bining data

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='UG', y='Tuition')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x='UG', y='Tuition', x_bins=5)
plt.show()
```
