```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/schoolimprovement2010grants.csv', index_col=0)
df.head()
```

## Stripplot

```python
# xem x√©t Award_Amount theo Model Selected
sns.stripplot(data=df, x='Award_Amount', y='Model Selected')
plt.show()
```

```python
sns.stripplot(data=df, x='Model Selected', y='Award_Amount')
plt.show()
```

```python
# xem x√©t Award_Amount theo Model Selected
sns.stripplot(data=df, x='Award_Amount', y='Model Selected', jitter=False)
plt.show()
```

## Swarmplot

```python
# xem x√©t Award_Amount theo Model Selected
sns.swarmplot(data=df, x='Award_Amount', y='Model Selected')
plt.show()
```

```python
sns.swarmplot(data=df, x='Award_Amount', y='Model Selected', hue='Region')
plt.show()
```
