```python
# xem bảng màu mặc định của seaborn
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

```python
# xem bảng màu deep6 của seaborn và tăng lên 10 ô màu
current_palette = sns.color_palette(palette='deep6', n_colors=10)
sns.palplot(current_palette)
plt.show()
```

```python
# xem tất cả các bảng màu của seaborn
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    current_palette = sns.color_palette()
    print(p)
    sns.palplot(current_palette)
    plt.show()
```

### Circular Color

```python
sns.palplot(sns.color_palette('dark6', 12))
```

### Sequential Color

```python
sns.palplot(sns.color_palette('Blues', 12))
plt.show()
```

### Diverging Color

```python
sns.palplot(sns.color_palette('coolwarm', 12))
plt.show()
```

```python
# dùng mã màu
sns.palplot(sns.diverging_palette(440, 40, n=10))
plt.show()
```

### áp dụng palette cho biểu đồ

```python
sns.palplot(sns.color_palette('Blues', 6))
plt.show()
```

```python
# xem xét sepal_length theo từng loại hoa (species)
myPalette = sns.color_palette('Blues', 6)
sns.swarmplot(data=df, x='species', y='sepal_length', palette=myPalette)
plt.show()
```

```python
# xem xét sepal_length theo từng loại hoa (species)
myPalette = sns.color_palette('Blues', 6)
sns.swarmplot(data=df, x='species', y='sepal_length', palette=myPalette)
plt.title("Iris's types", color='red')
plt.show()
```

```python
# xem xét sepal_length theo từng loại hoa (species)
myPalette = sns.color_palette('Blues', 6)
myplot = sns.swarmplot(data=df, x='species', y='sepal_length', palette=myPalette)
myplot.set_title("Iris's types", color='red')
plt.show()
```
