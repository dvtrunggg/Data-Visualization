```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/daily_show_guests_cleaned.csv')
df.head()
```

```python
df.info()
```

```python
table = pd.crosstab(df['Group'], df['YEAR'])
table
```

```python
# vẽ heatmap với các thông số mặc định
plt.figure(figsize=(8,8))
sns.heatmap(table)
# plt.xticks(rotation=45)
plt.show()
```

```python
# tùy chỉnh màu sắc cho heatmap
plt.figure(figsize=(8,8))
sns.heatmap(table, cmap='YlGnBu')
# plt.xticks(rotation=45)
plt.show()
```

```python
# tắt thanh chú thích màu sắc (color bar)
plt.figure(figsize=(8,8))
sns.heatmap(table, cmap='YlGnBu', cbar=False)
# plt.xticks(rotation=45)
plt.show()
```

```python
# tạo đường viền cho các ô trong heatmap
plt.figure(figsize=(10,8))
sns.heatmap(table, cmap='YlGnBu', linewidths=0.2)
plt.show()
```
