### xử lý các phần tử

```python
import math
seri2 = pd.Series([-4.4,5.8,2.2,-5.4,100,5.6,2.3,5,5,2.4])

seri2 **=2
seri2
```

```python
# dùng apply hoặc map đều ok, nhưng data frame thì chỉ có apply hoyy
seri2.apply(lambda x: x-2)
seri2

seri2.map(lambda x: x + 2)
seri2
```

```python
def XuLy(x):
    return x*2
```

```python
seri2 = seri2.map(lambda x: XuLy(x))

seri2 = seri2.apply(XuLy) #ko cần truyền tham số cũng dc:))
```

```python
#làm tròn cho đẹp hoy:))
ser2Test = seri2.round()

#thay đổi type của seri
ser2Test = ser2Test.astype('int32')
```

```python
def checkPrime(n):
    if n % 2 == 0 and n > 2: 
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True
```

```python
## dùng apply or map sẽ return true false
ser2Test.apply(checkPrime)
```

```python
serkq = ser2Test.map(checkPrime)
```

```python
ser2Test[ser2Test.apply(checkPrime)]

### serkq là 1 seri với value là true or false
seri2[serkq]
```
